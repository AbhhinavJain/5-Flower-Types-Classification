```python
import os
import cv2
from skimage.feature import hog
import numpy as np
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
X = []
labels = []
```

```python
image_dir1 = 'flower_images/Lilly/'
```

```python
for filename in os.listdir(image_dir1):
    
    img = cv2.imread(os.path.join(image_dir1, filename))
    resized_img = cv2.resize(img, (64, 64))
    X.append(resized_img)
    labels.append(1)
```

```python
image_dir2 = 'flower_images/Lotus/'
```

```python
for filename in os.listdir(image_dir2):
    
    img = cv2.imread(os.path.join(image_dir2, filename))
    resized_img = cv2.resize(img, (64, 64))
    X.append(resized_img)
    labels.append(2)
```

```python
image_dir3 = 'flower_images/Orchid/'
```

```python
for filename in os.listdir(image_dir3):
    
    img = cv2.imread(os.path.join(image_dir3, filename))
    resized_img = cv2.resize(img, (64, 64))
    X.append(resized_img)
    labels.append(3)
```

```python
image_dir4 = 'flower_images/Sunflower/'
```

```python
for filename in os.listdir(image_dir4):
    
    img = cv2.imread(os.path.join(image_dir4, filename))
    resized_img = cv2.resize(img, (64, 64))
    X.append(resized_img)
    labels.append(4)
```

```python
image_dir5 = 'flower_images/Tulip/'
```

```python
for filename in os.listdir(image_dir5):
    
    img = cv2.imread(os.path.join(image_dir5, filename))
    resized_img = cv2.resize(img, (64, 64))
    X.append(resized_img)
    labels.append(5)
```

```python
X = np.array(X)
labels = np.array(labels)
```

```python
X = X.reshape(len(X), -1)
```

```python
X.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=130)
pca.fit(X)
X_pca = pca.transform(X)
# print(X_pca)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, labels, test_size=0.25, random_state=0)
```

```python
X_train.shape
```

```python
# Define decision tree classifier as weak learner
dt = DecisionTreeClassifier(max_depth=5)
```

```python
# Define AdaBoost classifier with decision tree as base estimator
ada = AdaBoostClassifier(base_estimator=dt, n_estimators=200, learning_rate=0.5)
```

```python
# Train AdaBoost classifier
ada.fit(X_train, y_train)
```

```python

# Predict classes of test data
y_pred = ada.predict(X_test)
```

```python
# Calculate accuracy score of AdaBoost classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
